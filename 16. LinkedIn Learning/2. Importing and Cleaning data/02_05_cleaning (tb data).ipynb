## Python statistics essential training - 02_05_cleaning(tb data)

Standard imports

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
filename = r'F:\Data Analysis\Springboard\Data Science Career Track\16. LinkedIn Learning\Exercise Files\chapter2\02_05\tb.csv'

tb = pd.read_csv(filename)

tb
```

```python
#columns
tb.columns
```

```python
#melt the data
melted = tb.melt(['country','year'],['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],'sex_age','cases')
```

```python
melted.head()
```

```python
#Separating sex and age
melted['sex'] = melted['sex_age'].str.slice(0,1)
melted['age'] = melted['sex_age'].str.slice(1)
```

```python
melted.head()
```

```python
#We can map our age
melted['age'] = melted['age'].map({'04': '0-4', '514': '5-14', '1524': '15-24', '2534': '25-34', '3544': '35-44',
                                   '4554': '45-54', '5564': '55-64', '65': '65+', 'u': np.nan})
```

```python
melted['age']
```

```python
#drop missing values in cases column
final = melted.dropna(subset=['cases'])
```

```python
final.head()
```

```python
#Sort values
final.sort_values(['country','year','age','sex'])
final.head()
```

```python
#Select the columns we want
final = final[['country','year','age','sex','cases']]
final.head(10)
```

```python
#Export final to a new cvs file
final.to_csv('final.csv',index=False)
```
