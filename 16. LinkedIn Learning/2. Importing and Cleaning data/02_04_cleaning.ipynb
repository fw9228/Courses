```python
#melting data (subset of columns)
bshort = billboard[['artist.inverted','track','time','date.entered','x1st.week','x2nd.week','x3rd.week']]

bshort.head()
```

```python
#Renaming columns
bshort.columns = ['artist','track','time','date_entered','wk1','wk2','wk3']
bshort.head()
```

```python
bmelt = bshort.melt(['artist','track','time','date_entered'],['wk1','wk2','wk3'],'week','rank')
bmelt.head()
```

```python
#Getting the rank for a specific song
bmelt.query('track == "Liar"')
```

```python
bmelt.query('track == "Independent Women Part I"')
```

```python
#We can turn the weeks into number
bmelt['week'] = bmelt['week'].apply(lambda s: int(s[2]))
```

```python
bmelt.head()
```

```python
#Turn the dates into proper pandas date format
bmelt['date_entered'] = pd.to_datetime(bmelt['date_entered'])
```

```python
# date arithmetric. Getting time stamp
bmelt['date_entered'][0] + pd.Timedelta('7 days')
```

```python
#Adding an date column
bmelt['date'] = bmelt['date_entered'] + pd.Timedelta('7 days') * (bmelt['week'] - 1)
```

```python
bmelt.head()
```

```python
#Drop the previous date_entered
bmelt.drop(['date_entered'],axis=1,inplace=True)
```

```python
#Checking the song rank again
bmelt.query('track == "Liar"')
```

```python
#Sorting rank by artist and song
bfinal = bmelt[['artist','track','time','date','week','rank']]
bfinal.sort_values(['artist','track'],inplace=True)
```

```python
bfinal.head()
```

```python
#Drop duplicates
tracks = bfinal[['artist','track','time']].drop_duplicates()
tracks.head()
```

```python
#Naming the index
tracks.index.name = 'id'

#Resetting index
tracksid = tracks.reset_index()
tracksid.head()
```

```python
#merge the dataframes
pd.merge(tracksid,bfinal,on=['track','artist']).head()
```

```python
#The time is duplicated and have to be merged
pd.merge(tracksid,bfinal,on=['track','artist','time']).head()
```

```python
#tidying things up and droping the columns we dont need
tidy = pd.merge(tracksid,bfinal,on=['track','artist','time']).drop(['artist','track','time'],axis=1)
tidy.head()
```

```python
#Highest rank achieve by a song in week 1, 2, 3
tidy.loc[tidy[tidy.week == 1]['rank'].idxmin()]
```
