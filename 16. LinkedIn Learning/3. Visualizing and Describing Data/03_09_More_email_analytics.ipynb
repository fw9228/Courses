```python
file = r'F:\Data Analysis\Springboard\Data Science Career Track\16. LinkedIn Learning\Exercise Files\chapter3\03_09\mbox-cleaned.csv'

messages = pd.read_csv(file)
messages.head()
```

```python
messages['date'] = messages['date'].apply(lambda s: pd.to_datetime(s).tz_localize('UTC').tz_convert('America/Los_Angeles'))
```

```python
messages['dayofweek'] = pd.Categorical(messages['dayofweek'],
                                       ordered = True,
                                       categories = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
```

```python
days = messages.dayofweek.cat.categories
```

```python
for day in days:
    messages[messages.dayofweek == day].timeofday.plot(kind='density')
    
pp.legend(days)
```

```python
messages.groupby('dayofweek').timeofday.plot(kind='density')
pp.legend(days)
```
